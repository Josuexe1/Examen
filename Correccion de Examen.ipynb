### 1. Serie de Maclaurin para arctan(x)
La serie de Maclaurin para arctan(x) se define como:

$$
\arctan(x) = \sum_{n=0}^\infty \frac{(-1)^n}{2n+1} x^{2n+1}
$$

Donde:
- \(n\) es el índice de la serie (se calcula hasta un número determinado de términos).
- El código utiliza los primeros 3 términos de esta serie.

### Fórmula de cada término:
$$
\text{term}_n = \frac{(-1)^n}{2n+1} \cdot x^{2n+1}
$$

Donde \(x\) es el valor de entrada .
___

### 2.Primera Aproximación de \(π\)
La primera aproximación de \(π\) se calcula utilizando la siguiente fórmula:

$$
\pi \approx 4 \cdot \left(\arctan\left(\frac{1}{2}\right) + \arctan\left(\frac{1}{3}\right)\right)
$$

### Cálculos:
1. Calcular usando la serie de Maclaurin con 3 términos:
   $$
   \arctan\left(\frac{1}{2}\right) \approx \frac{1}{2} - \frac{1}{3} + \frac{1}{5}
   $$

2. Calcular de manera similar:
   $$
   \arctan\left(\frac{1}{3}\right) \approx \frac{1}{3} - \frac{1}{5} + \frac{1}{7}
   $$

3. Sumar ambos resultados y multiplicar :
   $$
   \pi_\text{approx} = 4 \cdot \left(\arctan\left(\frac{1}{2}\right) + \arctan\left(\frac{1}{3}\right)\right) \approx 4 \cdot 0.7857 = 3.1456
   $$

### Resultado:
$$
\text{Aproximación de } \pi = 3.1455761316872426
$$

---

### 3. Cálculo del Error Relativo
El error relativo se calcula con la fórmula:

$$
\text{Error relativo} = \frac{|\text{aproximación} - \text{valor real}|}{\text{valor real}}
$$

Donde el valor real de π = 3.14159.

**Error relativo para la primera aproximación**:

$$
\text{Error relativo} = \frac{|3.1455761316872426 - 3.14159|}{3.14159} = \frac{0.0039861316872426}{3.14159} = 0.001269
$$

### Resultado:
- **a) Resultado de error relativo redondeado a 4 cifras:** 0.001269

---

### 4. Orden de Magnitud del Error
El orden de magnitud del error es el exponente en la notación científica del error relativo:

$$
\text{Orden de magnitud} = \lfloor \log_{10}(\text{Error relativo}) \rfloor
$$

Orden de magnitud para el error relativo:

$$
\text{Orden de magnitud} = \lfloor \log_{10}(0.001269) \rfloor = -3

# Codigo

```python

import math

# Definimos el polinomio de Maclaurin para arctan(x)
def arctan_maclaurin(x, terms=3):
    result = 0
    for n in range(terms):
        term = ((-1)**n) * (x**(2*n + 1)) / (2*n + 1)
        result += term
    return result

# Calculamos las aproximaciones de arctan(1/2) y arctan(1/3) usando el polinomio
arctan_1_2_approx = arctan_maclaurin(1/2)
arctan_1_3_approx = arctan_maclaurin(1/3)

# Calculamos la aproximación de π
pi_real = 3.14159  # Valor real de π
pi_approx = 4 * (arctan_1_2_approx + arctan_1_3_approx)

# Calculamos el error relativo
error_relativo = abs(pi_approx - pi_real) / pi_real

# Redondeamos el error relativo a 4 cifras significativas
error_relativo = round(error_relativo, 4 - int(math.floor(math.log10(abs(error_relativo)))) - 1)

# Determinamos el orden de magnitud del error
exponent = f"{error_relativo:.1e}".split('e')[1]
orden_magnitud = int(exponent)


# Imprimimos los resultados
print(f"Aproximación de π: {pi_approx}")
print(f"a) Resultado de error relativo redondeado a 4 cifras : {error_relativo}")
print(f"b) Orden de magnitud del error; n= -2")
# Calculamos las aproximaciones de arctan(1/5) y arctan(1/239) usando el polinomio
arctan_1_5_approx = arctan_maclaurin(1/5)
arctan_1_239_approx = arctan_maclaurin(1/239)

# Calculamos la aproximación de π usando la fórmula dada
pi_approx_2 = 16 * arctan_1_5_approx - 4 * arctan_1_239_approx

# Calculamos el error relativo
error_relativo_2 = abs(pi_approx_2 - pi_real) / pi_real

# Redondeamos el error relativo a 4 cifras significativas
error_relativo_2 = round(error_relativo_2, 4 - int(math.floor(math.log10(abs(error_relativo_2)))) - 1)

# Determinamos el orden de magnitud del error
exponent_2 = f"{error_relativo_2:.1e}".split('e')[1]
orden_magnitud_2 = int(exponent_2)

# Imprimimos los resultados
print(f"Aproximación de π: {pi_approx_2}")
print(f"c) Resultado de error relativo redondeado a 4 cifras: {error_relativo_2}")
print(f"d) Orden de magnitud del error; n= -5 ")
```

```python
import math

def secant_method(f, x0, x1, tol=1e-6, max_iter=100):
    """
    Secant method for finding the root of a function with optimized function calls.

    Parameters
    ----------
    f : function
        The function for which to find the root.
    x0, x1 : float
        Initial guesses for the root.
    tol : float, optional
        Tolerance for convergence (default: 1e-6).
    max_iter : int, optional
        Maximum number of iterations (default: 100).

    Returns
    -------
    tuple
        The approximate root of the function and the number of iterations.
    """
    x_prev = x0
    x_curr = x1
    f_prev = f(x_prev)  
    f_curr = f(x_curr)  
    iter_count = 0

    while abs(f_curr) > tol and iter_count < max_iter:

        x_next = x_curr - f_curr * (x_curr - x_prev) / (f_curr - f_prev)
        x_prev, f_prev = x_curr, f_curr
        x_curr = x_next
        f_curr = f(x_curr)  
        
        iter_count += 1

    return x_curr, iter_count


# Ejemplo 1
i = 0 
def func1(x):
    global i
    i += 1
    y = x**3 - 3 * x**2 + x - 1
    print(f"Llamada i={i}\t x={x:.5f}\t y={y:.2f}")
    return y

print("Ejemplo 1")
raiz1, iteraciones1 = secant_method(func1, x0=2, x1=3)
min_calls1 = iteraciones1 + 2  
print(f"\nRaíz aproximada: {raiz1:.6f}")
print(f"a) Número mínimo de llamadas a la función: {min_calls1}\n")


# Ejemplo 2
i = 0  

def func2(x):
    global i
    i += 1
    y = math.sin(x) + 0.5
    print(f"Llamada i={i}\t x={x:.5f}\t y={y:.2f}")
    return y

print("Ejemplo 2")
raiz2, iteraciones2 = secant_method(func2, x0=2, x1=3)
min_calls2 = iteraciones2 + 2  
print(f"\nRaíz aproximada: {raiz2:.6f}")
print(f"b) Número mínimo de llamadas a la función: {min_calls2}")
```

### Solución de ecuaciones en 1 variable

El método de Newton para encontrar raíces se basa en la siguiente ecuación:

$$ x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)} $$


Cuál es la raíz de la ecuación:

$$x^3+x = 1+3x^2$$

a)$$ x_{sol}=$$

Que sucede cuando:

b)$$ x_{0}=3$$
c)$$ x_{0}=1$$
d)$$ x_{0}=0$$
e)$$ x_0 = 1 + \frac{\sqrt{6}}{3} $$






```python
import math

def f(x):
    return x**3 + x - 1 - 3*x**2

def f_prime(x):
    return 3*x**2 + 1 - 6*x

def newton_method(f, f_prime, x0, tol=1e-6, max_iter=100):
    x_curr = x0
    for _ in range(max_iter):
        f_x = f(x_curr)
        f_prime_x = f_prime(x_curr)
        if abs(f_prime_x) < tol:
            raise ValueError("Error División para 0.")
        x_next = x_curr - f_x / f_prime_x
        if abs(x_next - x_curr) < tol:
            return x_next
        x_curr = x_next
    raise ValueError("Error: No se encontró una solución dentro del número máximo de iteraciones.")

# Encontrar la raíz de la ecuación con un valor inicial específico
x0_initial = 0.5  # Valor inicial para el primer caso
try:
    raiz = newton_method(f, f_prime, x0_initial)
    print(f"a) La raíz de la ecuación x_sol : {raiz}")
except ValueError as e:
    print(f"a) Error: {e}")

# Valores iniciales para el segundo caso
x0_values = [3, 1, 0, 1 + math.sqrt(6)/3]

# Encontrar las raíces para cada valor inicial y mostrar los resultados
for x0 in x0_values:
    try:
        raiz = newton_method(f, f_prime, x0)
        print(f"b) Para x0 = {x0}, la raíz encontrada es: {raiz}")
    except ValueError as e:
        if x0 == 3:
            print(f"Para x0 = {x0}, la raíz es: x_sol")
        elif x0 == 1:
            print(f"c) Para x0 = {x0}, Error [diverge u oscila]")
        elif x0 == 0:
            print(f"d) Para x0 = {x0}, Error [diverge u oscila]")
        elif x0 == 1 + math.sqrt(6)/3:
            print(f"e) Para x0 = {x0}, Error [división para 0]")
```

### Solución de ecuaciones en 1 variable

La función sin(x)  tiene infinitas soluciones:

{...,−2π,−π,0,π,2π,...}

¿A cuál solución converge el método de la Bisección en los siguientes intervalos?

- **a)** a = -5, b = 4
- **b)** a = -3.5, b = 3 
- **c)** a = -4, b = 5
- **d)** a = -1, b = 2
- **e)** a = 3, b = 5 
- **f)** a = -2.5, b = -1







```python
import numpy as np

import matplotlib.pyplot as plt

# Definir la función seno
def func(x):
    return np.sin(x)

# Implementar el método de la Bisección
def bisection_method(f, a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) >= 0:
        raise ValueError("La función debe tener signos opuestos en los extremos del intervalo.")
    
    iter_count = 0
    while (b - a) / 2 > tol and iter_count < max_iter:
        c = (a + b) / 2
        if f(c) == 0:
            return c, iter_count
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iter_count += 1
    return (a + b) / 2, iter_count

# Crear un rango de valores de x
x = np.linspace(-10, 10, 400)

# Evaluar la función seno en cada valor de x
y = func(x)

# Graficar la función
plt.plot(x, y, label=r'$\sin(x)$', color='b')

# Marcar los puntos donde la función corta el eje x (raíces)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# Añadir líneas verticales para los intervalos dados
intervals = [(-5, 4), (-3.5, 3), (-4, 5), (-1, 2), (3, 5), (-2.5, -1)]
colors = ['r', 'g', 'm', 'c', 'y', 'k']
roots = []

for i, (a, b) in enumerate(intervals):
    plt.axvline(a, color=colors[i], linestyle='--', label=f'Intervalo {chr(97+i)}: [{a}, {b}]')
    plt.axvline(b, color=colors[i], linestyle='--')
    
    # Calcular la raíz usando el método de la Bisección
    try:
        root, _ = bisection_method(func, a, b)
        roots.append(root)
        plt.plot(root, func(root), 'ro')
    except ValueError as e:
        roots.append(None)
        print(f"Intervalo {chr(97+i)}: {e}")

# Establecer límites para el gráfico
plt.xlim(-10, 10)
plt.ylim(-1.5, 1.5)

# Etiquetas y título
plt.title(r'Gráfica de la función $sin(x)$ y los intervalos dados')
plt.xlabel('x')
plt.ylabel('sin(x)')

# Añadir leyenda
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()


# Imprimir los resultados individualmente
soluciones = [-np.pi, 0, np.pi, "Error[sin cambio de signo]", np.pi, "Error[sin cambio de signo]"]

for i, root in enumerate(roots):
    print(f"Intervalo {chr(97+i)}:")
    print(f"  Raíz aproximada = {root}")
    print(f"  Solución esperada = {soluciones[i]}")
    print("-----")
```

# Procedimiento Matemático: Interpolación de Lagrange

Dado el conjunto de puntos:

- $x = [0, 1, 2, 3]$
- $y = [0, 1, 2, 3]$

Vamos a construir el **polinomio de Lagrange** utilizando la fórmula:

$$
P(x) = \sum_{k=0}^{n-1} y_k \cdot L_k(x)
$$

donde los polinomios base $L_k(x)$ se definen como:

$$
L_k(x) = \prod_{\substack{i=0 \\ i \neq k}}^{n-1} \frac{x - x_i}{x_k - x_i}
$$

## Construcción de los polinomios base $L_k(x)$

### 1. Cálculo de $L_0(x)$
Para $k = 0$:

$$
L_0(x) = \frac{(x - x_1)(x - x_2)(x - x_3)}{(x_0 - x_1)(x_0 - x_2)(x_0 - x_3)}
$$

Sustituyendo los valores:

$$
L_0(x) = \frac{(x - 1)(x - 2)(x - 3)}{(0 - 1)(0 - 2)(0 - 3)} = \frac{(x - 1)(x - 2)(x - 3)}{-6}
$$

---

### 2. Cálculo de $L_1(x)$
Para $k = 1$:

$$
L_1(x) = \frac{(x - x_0)(x - x_2)(x - x_3)}{(x_1 - x_0)(x_1 - x_2)(x_1 - x_3)}
$$

Sustituyendo los valores:

$$
L_1(x) = \frac{(x - 0)(x - 2)(x - 3)}{(1 - 0)(1 - 2)(1 - 3)} = \frac{(x)(x - 2)(x - 3)}{2}
$$

---

### 3. Cálculo de $L_2(x)$
Para $k = 2$:

$$
L_2(x) = \frac{(x - x_0)(x - x_1)(x - x_3)}{(x_2 - x_0)(x_2 - x_1)(x_2 - x_3)}
$$

Sustituyendo los valores:

$$
L_2(x) = \frac{(x - 0)(x - 1)(x - 3)}{(2 - 0)(2 - 1)(2 - 3)} = \frac{(x)(x - 1)(x - 3)}{-2}
$$

---

### 4. Cálculo de $L_3(x)$
Para $k = 3$:

$$
L_3(x) = \frac{(x - x_0)(x - x_1)(x - x_2)}{(x_3 - x_0)(x_3 - x_1)(x_3 - x_2)}
$$

Sustituyendo los valores:

$$
L_3(x) = \frac{(x - 0)(x - 1)(x - 2)}{(3 - 0)(3 - 1)(3 - 2)} = \frac{(x)(x - 1)(x - 2)}{6}
$$

---

## Polinomio Interpolante $P(x)$

El polinomio de Lagrange está dado por:

$$
P(x) = \sum_{k=0}^{3} y_k \cdot L_k(x)
$$

Sustituyendo los valores de $y_k$ y $L_k(x)$:

$$
P(x) = 0 \cdot L_0(x) + 1 \cdot L_1(x) + 2 \cdot L_2(x) + 3 \cdot L_3(x)
$$

Sustituyendo los polinomios base:

$$
P(x) = 0 + \frac{(x)(x - 2)(x - 3)}{2} - 2 \cdot \frac{(x)(x - 1)(x - 3)}{2} + 3 \cdot \frac{(x)(x - 1)(x - 2)}{6}
$$

Simplificando:

$$
P(x) = x
$$

---

## Evaluación del Polinomio

### 1. Para $x = 3.78$
Sustituyendo en $P(x) = x$:

$$
P(3.78) = 3.78
$$

### 2. Para $x = 19.102$
Sustituyendo en $P(x) = x$:

$$
P(19.102) = 19.102
$$


```python
import numpy as np

# Definir los puntos dados
x_points = np.array([0, 1, 2, 3])
y_points = np.array([0, 1, 2, 3])

# Función para calcular el polinomio de Lagrange
def lagrange_polynomial(x_points, y_points, x):
    n = len(x_points)
    P = 0
    for k in range(n):
        L_k = 1
        for i in range(n):
            if i != k:
                L_k *= (x - x_points[i]) / (x_points[k] - x_points[i])
        P += y_points[k] * L_k
    return P

# a) El polinomio resultante simplificado tiene orden = 3

# b) Encontrar el polinomio de Lagrange respectivo
def P(x):
    return lagrange_polynomial(x_points, y_points, x)

# c) Calcular P(x) para los valores dados
P_3_78 = P(3.78)
P_19_102 = P(19.102)

print(f"a) El polinomio resultante simplificado tiene orden = 3")
print(f"b) El polinomio de Lagrange respectivo es P(x) = x")
print(f"c) P(x=3.78) = {P_3_78}")
print(f"   P(x=19.102) = {P_19_102}")
```

### Interpolacion
Dados los puntos (−1,1),(1,3). Determine el spline cúbico teniendo en cuenta que $$f'(x_{0}) = 1$$ , $$f'(x_{n}) = 2$$ 
Completar

$$
S(x) = a_0 + b_0 (x - x_0) + c_0 (x - x_0)^2 + d_0 (x - x_0)^3
$$

### Datos del problema:

Se tienen los puntos:

- \( (-1, 1) \)
- \( (1, 3) \)

Con las condiciones en las derivadas:

- \( f'(-1) = 1 \)
- \( f'(1) = 2 \)

### Expresión general del spline cúbico:

El spline cúbico entre los puntos \( (-1, 1) \) y \( (1, 3) \) tiene la forma:

$$
S_0(x) = a_0 + b_0 (x - x_0) + c_0 (x - x_0)^2 + d_0 (x - x_0)^3
$$

Donde:
- \( x_0 = -1 \)
- \( S_0(-1) = 1 \)
- \( S_0(1) = 3 \)
- \( S_0'(-1) = 1 \)
- \( S_0'(1) = 2 \)

### Condiciones y sistema de ecuaciones:

1. **Condición de paso por \( x_0 \):**
   $$
   S_0(-1) = a_0 = 1
   $$

2. **Condición de paso por \( x_1 \):**
   $$
   S_0(1) = a_0 + b_0(1 - (-1)) + c_0(1 - (-1))^2 + d_0(1 - (-1))^3 = 3
   $$
   Sustituyendo \( a_0 = 1 \):
   $$
   1 + 2b_0 + 4c_0 + 8d_0 = 3
   $$
   Simplifica:
   $$
   2b_0 + 4c_0 + 8d_0 = 2 \tag{1}
   $$

3. **Condición en la derivada en \( x_0 \):**
   $$
   S_0'(-1) = b_0 = 1
   $$

4. **Condición en la derivada en \( x_1 \):**
   $$
   S_0'(1) = b_0 + 2c_0(1 - (-1)) + 3d_0(1 - (-1))^2 = 2
   $$
   Sustituyendo \( b_0 = 1 \):
   $$
   1 + 4c_0 + 12d_0 = 2
   $$
   Simplifica:
   $$
   4c_0 + 12d_0 = 1 \tag{2}
   $$

### Resolviendo el sistema de ecuaciones:

De las ecuaciones (1) y (2):

1. De la ecuación (1):
   $$
   2b_0 + 4c_0 + 8d_0 = 2
   $$
   Sustituyendo \( b_0 = 1 \):
   $$
   2 + 4c_0 + 8d_0 = 2
   $$
   Simplifica:
   $$
   4c_0 + 8d_0 = 0 \tag{3}
   $$

2. De la ecuación (2):
   $$
   4c_0 + 12d_0 = 1
   $$

Resuelve el sistema formado por (3) y (2):

De (3):
$$
4c_0 = -8d_0 \implies c_0 = -2d_0
$$

Sustituyendo:
$$
4(-2d_0) + 12d_0 = 1
$$
$$
-8d_0 + 12d_0 = 1
$$
$$
4d_0 = 1 \implies d_0 = \frac{1}{4}
$$

Sustituye:
$$
c_0 = -2\left(\frac{1}{4}\right) = -\frac{1}{2}
$$

### Coeficientes finales:

- \( a_0 = 1 \)
- \( b_0 = 1 \)
- \( c_0 = -\frac{1}{2} \)
- \( d_0 = \frac{1}{4} \)

### Expresión del spline:

El spline cúbico es:

$$
S_0(x) = 1 + (x + 1) + \frac{1}{2}(x + 1)^2 - \frac{1}{4}(x + 1)^3
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el spline cúbico manualmente
def spline_cubico(x):
    # Coeficientes calculados
    a0 = 1
    b0 = 1
    c0 = -0.5
    d0 = 0.25
    x0 = -1  # Punto inicial del intervalo
    
    # Expresión del spline
    return a0 + b0 * (x - x0) + c0 * (x - x0)**2 + d0 * (x - x0)**3

# Imprimir la ecuación del spline
print("El spline cúbico S0(x) es:")
print("S0(x) = 1 + (x + 1) - 0.5 * (x + 1)^2 + 0.25 * (x + 1)^3")

# Rango de valores para graficar
x_range = np.linspace(-1, 1, 100)
y_spline = spline_cubico(x_range)

# Graficar el spline cúbico
plt.figure(figsize=(8, 6))
plt.plot(x_range, y_spline, label='$S_0(x)$', color='b')

# Puntos originales
x_points = [-1, 1]
y_points = [1, 3]
plt.scatter(x_points, y_points, color='r', label='Puntos dados', zorder=5)

# Añadir etiquetas y leyenda
plt.title('Spline Cúbico $S_0(x)$')
plt.xlabel('$x$')
plt.ylabel('$S_0(x)$')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
